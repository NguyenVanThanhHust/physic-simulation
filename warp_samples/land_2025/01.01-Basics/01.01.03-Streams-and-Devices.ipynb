```python
import warp as wp

print(wp.get_devices())
```

```python
# Allocate test_array_0 on the default device
test_array_0 = wp.zeros(10, dtype=wp.float32)
print(f"{test_array_0.device=}")

# Allocate test_array_1 on the cuda:1 device
test_array_1 = wp.zeros(10, dtype=wp.float32, device='cuda:0')
print(f"{test_array_1.device=}")
```

```python
# Allocate test_array_0 on the cuda:0
with wp.ScopedDevice('cuda:0'):
    test_array_0 = wp.zeros(10, dtype=wp.float32)
print(f"{test_array_0.device=}")

# Allocate test_array_1 on the cuda:1 device
with wp.ScopedDevice('cuda:0'):
    test_array_1 = wp.zeros(10, dtype=wp.float32)
print(f"{test_array_1.device=}")
```
